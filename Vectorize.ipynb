```python
from SymEigen import *
from sympy import Trace

F = Eigen.Matrix('F', 3, 3)
VecF = F.Vectorize('VecF')
Ic =  Trace(F.T*F)
```

```python
dIcdVecF = VecDiff(Ic, VecF)
dIcdVecF
```

```python
VecF
```

```python
ddICddVecF = VecDiff(dIcdVecF, VecF)
ddICddVecF
```

```python
Gen = EigenFunctionGenerator()
Closure = Gen.Closure(VecF)
print(Closure('dIcdVecF', dIcdVecF))
```

Just see the difference between the Vectorized Output and the Non-Vectorized Output, they are both Ok.

```python
Closure = Gen.Closure(F)
print(Closure('dIcdF', dIcdVecF))
```
